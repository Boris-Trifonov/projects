# Исследование продаж компьютерных игр

**Входные данные:**

- Данные с о продажах компьютергных игр в интернет-магазине «Стримчик» за несколько лет:

https://code.s3.yandex.net/datasets/games.csv
        
**Цель исследования:**
        
- Выявить закономерности, определяющие успешность игры.

- Возможность спланировать рекламные кампании на будущий год, на основе исследования.
        
**Задачи**

- Проверить гипотезы:

    1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
    
    2. Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.


**Этапы исследования:**

- Обзор данных.
- Предобработка данных.
- Исследовательский анализ данных.
- Сформировать портрет пользователя каждого региона.
- Проверка гипотез.
- Общий вывод.

**Описание данных**

`Name` — название игры

`Platform` — платформа

`Year_of_Release` — год выпуска

`Genre` — жанр игры

`NA_sales` — продажи в Северной Америке (миллионы проданных копий)

`EU_sales` — продажи в Европе (миллионы проданных копий)

`nan` - пропуски. С ними понятно. 
Значение, либо не присвоено, либо не требовалось присваивать на момент выпуска игры, либо просто данные отсутствуют или пропущены.
До 1994 года данный рейтинг вообще не существовал, и долгое время был необязателен. Во всем виноват "Mortal Kombat" - именно с его жестокости все и началось 

`EC` («Early childhood») — «Для детей младшего возраста»: Игра подходит для детей от 3 лет

`E` («Everyone») — «Для всех»: Содержание вполне подходит для возрастной категории от 6 лет.

`E10+` («Everyone 10 and older») — «Для всех от 10 лет и старше»

`T` («Teen») — «Подросткам»: Игра подходит для лиц от 13 лет.

`M` («Mature») — «Для взрослых»: Материалы игры не подходят для лиц младше 17 лет.

`AO` («Adults Only 18+») — «Только для взрослых»: Содержание игры только для взрослых старше 18 лет.

`RP` («Rating Pending») — «Рейтинг ожидается»: Продукт был отправлен в ESRB и ожидает присвоения рейтинга. 

Данные взяты: https://ru.wikipedia.org/wiki/Entertainment_Software_Rating_Board

`K-A` С этим рейтингом сложнее. Он был в первоначальной версии ESRB. в 1998 году был переименован в Е – для всех возрастов, а также чуть позже был добавлен рейтинг Е10+, как промежуточный между Е и Т.

Пока подержим эти данные в уме. Если будем исследовать игры по рейтингу значения `RP` можно будет прировнять к пропускам, а рейтинг `K-A` либо проигнориируем, либо прировняем к `E10+`.

## Исследовательский анализ данных

### Выпуск игр по годам
Сгруппируем данные и посмотрим сколько игр выпускалось в разные годы.

```python
year_fitered = data.groupby('year_of_release').agg(games=('name', 'count')).reset_index()
year_fitered
```

### Влияние отзывов на продажи игр всех потенциальных платформ

```python
for name in data_potential['platform'].unique():
    platform = data_potential[data_potential['platform'] == name]
    
    fig, ax = plt.subplots(1 ,2, figsize=(15,5))
    sns.scatterplot(data=platform, x='user_score', y='total_sales', ax=ax[0])
    plt.title('Влияние отзывов критиков на продажи игр')
    sns.scatterplot(data=platform, x='critic_score', y='total_sales', ax=ax[1])
    plt.title('Влияние отзывов пользователей на продажи игр')
    fig.suptitle(name, fontsize=15)
    ax[0].set(xlabel='Оценки игроков')
    ax[1].set(xlabel='Оценки критиков')
    ax[0].set(ylabel='Количество копий,млн')
    ax[1].set(ylabel='Количество копий,млн')
    plt.show()
    
    critic_corr = platform.critic_score.corr(platform.total_sales).round(2)
    print('Корреляция оценок критиков и продаж:', critic_corr)
    
    user_corr = platform.user_score.corr(platform.total_sales).round(2)
    print('Корреляция оценок игроков и продаж:', user_corr)
    
    print('--------------------------------------------------------------------------------------------------------------')
```

**Промежуточный вывод**

Связь между продажами и оценками критиков для всех платформ, является прямой, слабой степени выраженности. Коэффициент корреляции: от 0.17 до 0.43.

Корреляция между продажами и оценками игроков для всех платформ, кроме 3DS, отсутствует.
Связь между продажами и оценками игроков для платформы 3DS, является прямой, слабой степени выраженности. Коэффициент корреляции: от 0.22

## Проверка гипотез

### Средние пользовательские рейтинги

Формулируем гипотезы:

H0 = Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

H1 = Средние пользовательские рейтинги платформ Xbox One и PC разные.

```python
XOne_user_score = data[data.platform == 'XOne']['user_score']
PC_user_score = data[data.platform == 'PC']['user_score']

results = st.ttest_ind(XOne_user_score, PC_user_score, equal_var = False)

alpha = 0.01

print('p-value:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Получили значение p-value: nan. Попробуем проигнорировать пропуски - удалим эти строки для теста

```python
XOne_user_score = data[data.platform == 'XOne']['user_score']
PC_user_score = data[data.platform == 'PC']['user_score']

results = st.ttest_ind(XOne_user_score.dropna(), PC_user_score.dropna(), equal_var = False)

alpha = 0.01

print('p-value:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Очень высокий показатель p-value, говорит о том, что очень высокая вероятность получить различия между выборками.

У нас достаточно оснований, чтобы отвергнуть H0 и утверждать, что средние пользовательские рейтинги платформ Xbox One и PC значимо различаются

### Средние пользовательские рейтинги жанров Action и Sports

Формулируем гипотезы:

H0 = Средние пользовательские рейтинги жанров Action и Sports одинаковые.

H1 = Средние пользовательские рейтинги жанров Action и Sports разные.

```python
Action_user_score = data[data.genre == 'Action']['user_score']
Sports_user_score = data[data.genre == 'Sports']['user_score']

results = st.ttest_ind(Action_user_score.dropna(), Sports_user_score.dropna(), equal_var = False)

alpha = 0.01

print('p-value:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

У нас нет оснований утверждать, что средние пользовательские рейтинги жанров Action и Sports значимо различаются».
